```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

```python
mpg['total'].describe()
```

```python
mpg['total'].plot.hist()
```

<p> total 연비의 평균과 중앙값이 약 20이다</p>
<p> total 연비가 20 ~ 22.5 사이인 자동차 모델이 많다.</p>
<p>연비는 대부분 25 이하, 25를 넘기는 자동차 모델은 많지 않다.</p> 

<p>연비가 20이 넘는 자동차에 고연비 합격 판정을 내리면 될듯하다.</p>

```python
mpg['test'] = np.where(mpg['total'] >= 20, "pass", "fail")
mpg.head()
```

```python
count_test = mpg['test'].value_counts()
```

```python
count_test.plot.bar(rot = 0)
```

```python
mpg['Grade'] = np.where(mpg['total'] >= 30, 'A',
                       np.where(mpg['total'] >= 20, 'B', 'C'))
```

```python
count_grade = mpg['Grade'].value_counts().sort_index()
count_grade
```

```python
nmpg.loc[[9,13,57,92], 'drv'] = 'k'
nmpg.loc[[28,42,128,202], 'cty'] = [3,4,39,42]
```

```python
nmpg['drv'].value_counts()
```

```python
nmpg['drv'] = np.where(nmpg['drv'].isin(['f', '4', 'r']), nmpg['drv'], np.nan)
```

```python
sns.boxplot(data = nmpg, y= 'cty')
```

```python
pct25=nmpg['cty'].quantile(.25)
pct75= nmpg['cty'].quantile(.75)
```

```python
iqr = pct75 - pct25
high = pct75 + iqr*1.5
low = pct25 - iqr*1.5
```

```python
nmpg['cty'] = np.where((nmpg['cty'] > high) | (nmpg['cty'] < low) , np.nan, nmpg['cty'])
```

```python
sns.boxplot(data = nmpg, y= 'cty')
```
